Ahora separamos solo los valores numericos del codigo para poder filtrarlo.

```python
def obtenerN(x):
    x = ''.join(c for c in x if c.isdigit())
    return int(x)

reporteNoElimS['Código Suiza'] = reporteNoElimS['Código Suiza'].map(obtenerN)
reporteNoElimS
```
