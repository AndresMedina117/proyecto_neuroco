```python
hoja1 = hipocampo['amygdalar-nuclei lh']
hoja1
```

```python
sans = hoja1[hoja1['Measure:volume'].str.startswith('S')]
sans['Measure:volume'] = sans['Measure:volume'].map(lambda x: x.split('/')[0])
print(sans.shape)
sans.head()
```

Generamos los cambios con lo que hemos encontrado hasta ahora, estos cambios generados no afectarán a los pacientes Suiza.

```python
cambios = open('cambios.txt', 'w')
```

```python
for i in range(sans.shape[0]):
    fila = sans.iloc[i]
    subject = fila['Measure:volume']
        
    result = resumen[resumen.CODIGO == subject]
    if result.size == 5:
        grupo = result.GRUPO.to_numpy()[0]
        sans.iloc[i,0] = grupo
        cambios.write(f'{subject}|{grupo}\n')
```

```python
np.unique(sans['Measure:volume'], return_counts = True)
```

Como vemos todos los pacientes que quedaron fueron 2 suiza y 2 SAN que NO aparecen en la BD.

Ahora podemos trabajar con la mayoria de los que quedan, que son los pacientes Suiza.

```python
preprocesamiento = reporte['preprocesamiento'].iloc[:,:4].copy()
preprocesamiento = preprocesamiento.fillna('')
preprocesamiento = preprocesamiento[preprocesamiento['Código Suiza'] != '']
preprocesamiento
```

```python
for i in range(sans.shape[0]):
    fila = sans.iloc[i]
    subject = fila['Measure:volume']
    subjectC = ''.join(char for char in subject if char.isdigit())
    if not subjectC:
        continue
    subjectC = int(subjectC)

    busqueda = preprocesamiento[preprocesamiento['Código Suiza'].str.endswith(str(subjectC))]
    if busqueda.size == 0: 
        continue
    busqueda = busqueda['Código Biomarcadores'].to_numpy()[0]
        
    result = resumen[resumen.CODIGO == busqueda]
    if result.size == 5:
        grupo = result.GRUPO.to_numpy()[0]
        sans.iloc[i,0] = grupo
        cambios.write(f'{subject}|{grupo}\n')

cambios.close()
```
